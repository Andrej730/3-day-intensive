# 🔎 Знакомство с Python

###### Типы переменных

Главным объектом в Data Science, конечно, являются данные. С точки зрения python все данные "помещаются" в переменные - это некоторое имя, по которому можно обратиться к определённой области памяти в компьютере, где эти данные хранятся. Имя переменной может быть любым, главное, чтобы оно не начиналось с цифры. Также нужно отметить, что в переменные можно положить объекты разных типов (при этом, в отличие от, например, языка С, С++ указывать тип вручную не нужно, он определится автоматически).

Давайте попробуем создать несколько переменных разных типов:

```python
# создаём переменную-число
number = 5
# переменную-строку
string = 'name'
# переменную-индикатор (булеву)
flag = True
# переменную-массив даных
array = [number, string, flag]
```
